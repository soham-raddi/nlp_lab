```python
corpus = """
natural language processing enables computers to understand natural language
natural language processing is used in many applications
"""
```

```python
tokens = corpus.lower().split()
print(tokens)
```

```python
from collections import defaultdict
def build_ngram_model(tokens, n):
    ngram_model = defaultdict(lambda: defaultdict(int))
    for i in range(len(tokens) - n):
        context = tuple(tokens[i:i+n-1])
        next_word = tokens[i+n-1]
        ngram_model[context][next_word] += 1
    return ngram_model
```

```python
n = 2
ngram_model = build_ngram_model(tokens, n)
```

```python
def predict_next_word(ngram_model, context):
    context = tuple(context.lower().split())
    if context in ngram_model:
        next_words = ngram_model[context]
        predicted_word = max(next_words, key=next_words.get)
        return predicted_word
    else:
        return None
```

```python
context = "natural"
predicted_word = predict_next_word(ngram_model, context)
print(f"Predicted next word for context '{context}': {predicted_word}")
```
