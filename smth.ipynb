Constituency Parsing

```python
import nltk
from nltk import CFG
from nltk.parse import ChartParser

#Create CFG (Context-Free Grammar)
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N | Det Adj N | 'I' | 'She'
VP -> V NP | V
Det -> 'a' | 'the'
Adj -> 'big' | 'small'
N -> 'dog' | 'cat' | 'park'
V -> 'saw' | 'ran'
""")

#Create Parser
parser = ChartParser(grammar)

#Input Sentence
sentence = "I saw a dog"
tokens = sentence.split()

#Parse Sentence and Print Tree
print("\nParse Trees:\n")

try:
    for tree in parser.parse(tokens):
        print(tree)
        tree.pretty_print()
except ValueError:
    print("Sentence does not match the grammar.")
```

Shallow Parsing

```python
import nltk

# 1. Download required NLTK resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
def shallow_parse(sentence):
    # 2. Tokenize the sentence into words
    tokens = nltk.word_tokenize(sentence)

    # 3. Assign Part-of-Speech (POS) tags to each word
    tagged = nltk.pos_tag(tokens)

    # 4. Define Chunk Grammar (e.g., Noun Phrase: optional Determiner + Adjectives + Noun)
    # Grammar rules use RegEx-like syntax for POS tags
    grammar = r"NP: {<DT>?<JJ>*<NN>}"

    # 5. Create a Regex Parser and parse the tagged sentence
    cp = nltk.RegexpParser(grammar)
    result = cp.parse(tagged)

    return result
```

```python
# Example sentence
text = "The quick brown fox jumps over the lazy dog."
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
parsed_tree = shallow_parse(text)
```

```python
parsed_tree = shallow_parse(text)
```

```python
# 6. Output the result
print(parsed_tree)
# To visualize the tree in a separate window (requires Ghostscript/Tkinter)
# parsed_tree.draw()
```

Dependency Parsing

```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Input sentence
doc = nlp("The cat sat on the mat")

print("Token\tDependency\tHead")

for token in doc:
    print(f"{token.text}\t{token.dep_}\t\t{token.head.text}")
```

```python
from spacy import displacy
displacy.serve(doc, style="dep")
```
